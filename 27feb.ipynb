```python
class Grandfather:
    def house(self):
        print("owns a house")
class Father(Grandfather):
    def car(self):
        print("owns a car")
class Son(Father):
    def bike(self):
        print("owns a bike")
class Grandson(Son):
    def cycle(self):
        print("owns a cycle")

g =Grandson()
print("Grandson can access every class")
g.house()
g.car()
g.bike()
g.cycle()
```
