```python
class Student:
    def __init__(self,name,age,grade):#constructor
        self.name = name
        self.age = age
        self.grade = grade

    def display(self):
        print("name:", self.name)
        print("age:",self.age)
        print("grade:",self.grade)

s1 = Student("Ali",20,"A grade")
s2 = Student("Ahmed",25,"Bgrade ")
s1.display()
print()
s2.display()

```


# Inheritance

```python
class Person:

    def __init__(self, name,cast):
        self.name = name
        self.cast= cast
      
    def display(self):
        print("Name:", self.name)
        print("Name:", self.cast)

class Teacher(Person):
    def __init__(self,name,cast,subject):
        super().__init__(name,cast)
        self.subject = subject
    def Info(self):
        print(f"my name is {self.name}.I am {self.cast}.My subject is {self.subject}")

n0 = Person("Ali","Abbasi")
n1 = Teacher("Ali","Abbasi","Math")
n2 = Teacher("Bilal","Kiani","A.I")
print(n0.name)
n0.display()
n1.Info()
n2.Info()
```
