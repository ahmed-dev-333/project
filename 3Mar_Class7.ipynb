```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis


data = [1,22,30,49,56,64,72,81,98,0]

mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)
cv = (std_dev / mean) * 100
skewness = skew(data)
kurt = kurtosis(data)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Coefficient of Variation:", cv)
print("Skewness:", skewness)
print("Kurtosis:", kurt)

# Histogram
plt.hist(data, bins=10)
plt.title("Histogram of Data")
plt.show()
```

```python
data = np.random.randint(100, 1000, 100)

mean = np.mean(data)
std = np.std(data)

max_value = np.max(data)
min_value = np.min(data)
z_score = (max_value - mean) / std
skewness = skew(data)

print("Dataset:", data)
print("Largest no:",max_value)
print("Smallest no:",min_value)
print("Mean:", mean)
print("Standard Deviation:", std)
print("Z-score of max value:", z_score)
print("Skewness:", skewness)

# Percentiles
print("25th Percentile:", np.percentile(data, 25))
print("50th Percentile:", np.percentile(data, 50))
print("75th Percentile:", np.percentile(data, 75))

# Boxplot
plt.boxplot(data)
plt.title("Boxplot")
plt.show()

# Histogram
plt.hist(data, bins=100)
plt.title("Histogram of Data")
plt.show()
```
