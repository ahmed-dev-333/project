```python
import numpy as np
import pandas as pd
import statistics as stats
age = [3,4,6,2,4,8,6,1,2,9]
print("List of age is:", age)
print ("Mean:", np.mean(age))
print ("Median:", np.median(age))
print("Mode:",stats.mode(age))
```

# USING DATAFRAME

```python
data = {
    "Name":["A","B","C","D","E","F","G","H","I","J"],
    "Age":[11,11,33,33,11,66,77,88,99,00]
}
t = pd.DataFrame(data)
print(t)
print("MEAN OF AGES:",t["Age"].mean())
print("MEDIAN OF AGES:",t["Age"].median())
print("MODE OF AGES: \n", t["Age"].mode())
```
